```python
import pandas as pd
import os
```

```python
files = [file for file in os.listdir("./Pandas-Data-Science-Tasks-master/SalesAnalysis/Sales_Data")]
all_data = pd.DataFrame()
for file in files:
    df = pd.read_csv("./Pandas-Data-Science-Tasks-master/SalesAnalysis/Sales_Data/"+file)
    all_data = pd.concat([all_data, df])
# all_data.to_csv("all_data.csv", index = False)
```

```python
all_data.head()
```

```python
all_data = all_data.dropna(how = "all")
all_data = all_data[all_data["Order Date"].str[0:2] != "Or"]
all_data["Month"] = all_data["Order Date"].str[0:2]
all_data["Month"] = all_data["Month"].astype("int32")
all_data["Price Each"] = pd.to_numeric(all_data["Price Each"])
all_data["Quantity Ordered"] = pd.to_numeric(all_data["Quantity Ordered"])
all_data.head()
```

```python
all_data.reset_index(drop=True, inplace=True)
all_data.head()
```

## SALES COLUMN

```python
all_data["Sales"] = all_data["Quantity Ordered"] * all_data["Price Each"]
all_data.head()
```

## SALES 

```python
all_data.groupby("Month").sum()["Sales"]
```

## CITY COLUMN

```python
all_data["City"] = all_data["Purchase Address"].apply(lambda x: x.split(",")[1])
all_data.head()
```

## REVENUE BY CITY

```python
all_data.groupby("City").sum()["Sales"]
```

```python
all_data.groupby("Product").sum()["Sales"]
```

## PRODUCTS SOLD

```python
all_data.groupby("Product").sum()["Quantity Ordered"]
```

```python
all_data.to_csv("final_data.csv", index = False)
```
